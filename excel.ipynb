```python
import matplotlib.pyplot as plt
import japanize_matplotlib
#import openpyxl
from openpyxl import load_workbook
import os

# 現在のディレクトリを取得
current_directory = os.getcwd()

# excelフォルダのパスを作成
folder_path = os.path.join(current_directory, 'excel')

# フォルダ内のExcelファイルをリストアップ
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

plt.figure(figsize=(8, 6))

# 各Excelファイルを読み込む
for file_name in excel_files:
    file_path = os.path.join(folder_path, file_name)
    wb = load_workbook(file_path)

    sheet = wb.active

    x_data = []
    y_data = []
    random_data = []
    for row in sheet.iter_rows(values_only=True):
        x_data.append(row[0])
        y_data.append(row[2])
        random_data.append(row[1])

    #plt.plot(x_data[2:41], y_data[2:41], label=y_data[0])
    plt.plot(x_data[2:], y_data[2:], label=y_data[0])

    wb.close()

#plt.plot(x_data[2:41], random_data[2:41], label=random_data[0])
plt.plot(x_data[2:], random_data[2:], label=random_data[0])

problem = 'ZDT4'

plt.xlabel('世代数')
plt.ylabel('HV')
plt.title(problem)
plt.legend()
#plt.grid(True)

plt.savefig(f'{problem}_MOEAD_HV.png')

wb.close()
    
    
```
